```python
import requests
import matplotlib.pyplot as plt
import time 
```

```python
lats = []
longs = []
times = []

for i in range(100):
    response = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    if response.status_code == 200:
        data = response.json()
        lats.append(data['latitude'])
        longs.append(data['longitude'])
        times.append(data['timestamp'])
    else: print("Failed to fetch data.")
    time.sleep(10)
```

```python
plt.figure(figsize=(10,10))
plt.plot(longs,lats,marker="o",color="red")
plt.title("ISS Path")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
